```python
import os
import MnistSet
from MnistSet import MNISTSet
from MnistSet import get_loader
```

```python
dataset_mnist_set = MNISTSet(full=True)
```

```python
data_loader  = get_loader(dataset_mnist_set,batch_size=32)
```

```python
data_loader
```

```python
data_iter = iter(data_loader)
```

```python
label,set, mask = data_iter.next()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.grid(False)
plt.tight_layout()
for idx, s in enumerate(set):
    x = s[1, :]
    y = s[0, :]
    plt.scatter(x, y,color ='b')
    plt.axis([0, 1, 1, 0])
    plt.title('Label: {}'.format(label[idx]))
    #plt.savefig('test.png')
    plt.show()
```
